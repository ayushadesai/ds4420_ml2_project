# NYC Airbnb — Data Preprocessing

Loads the raw Inside Airbnb November 2025 scrape (`Nov_data.csv`), audits missing values across all 79 columns, drops columns with more than 15,000 missing values, and saves the result to `airbnb_clean.csv` for use in both the content-based filtering notebook and the Bayesian models.

```python
import pandas as pd
import numpy as np
```

```python
airbnb = pd.read_csv('Nov_data.csv')
```

```python
airbnb
```

```python
missing_val_list = []

for values in airbnb:
    print(values, airbnb[values].isnull().sum())
    missing_val_list.append(airbnb[values].isnull().sum())
```

```python
# Drops columns where more than 15,000 values are missing
airbnb = airbnb.dropna(axis=1, thresh=len(airbnb) - 15000)
```

```python
airbnb.to_csv('airbnb_clean.csv')
```
